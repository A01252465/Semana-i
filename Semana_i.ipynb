```python
A = pd.read_csv('/content/drive/MyDrive/Gasto-Agua.csv')
```

```python
y = A['Dentro de vivienda'].values.reshape(-1, 1) #dependant 
x = A['Domestico'].values.reshape(-1, 1) #independant

lm = LinearRegression().fit(x,y)
lmp = lm.predict(x)

plt.scatter(x, y)
plt.plot(x, lmp, color='red')
plt.show()
```
